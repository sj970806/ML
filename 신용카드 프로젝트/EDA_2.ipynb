## oversampling

```python
train = pd.read_csv('../data/신용카드/train_ver1.csv',sep=",", encoding = 'cp949', index_col = 0)
train.head()
```

```python
test = pd.read_csv('../data/신용카드/train_ver1.csv',sep=",", encoding = 'cp949', index_col = 0)
test.head()
```

```python
train.shape
```

```python
X_train = train.iloc[:,:-1]
y_train = train['credit']
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state = 13)
X_train_over, y_train_over = smote.fit_resample(X_train, y_train)
```

```python
print('Before over sampling : ', np.unique(y_train, return_counts = True))
print('After over sampling : ', np.unique(y_train_over, return_counts = True))
```

```python
train_over = pd.concat([X_train_over,y_train_over], axis = 1)
train_over.shape
```

```python
train_over.to_csv('../data/신용카드/train_oversample_ver1.csv',sep=",", encoding = 'cp949')
```
